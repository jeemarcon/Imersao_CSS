```python
mapa = {'droga':'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
dados.head()
```

```python
import seaborn as sns

sns.countplot(x = 'composto', data=dados)
```

```python
compostos = dados['composto'].value_counts()[0:5]
```

```python
compostos.index ##apenas o nome dos compostos
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
```

```python
cod_compostos
```

```python
compostos_top = dados.query('composto in @cod_compostos')
compostos_top
```

```python
sns.set() ###Deixar o gráfico no padrão seaborn 
plt.figure(figsize=(10, 8))
gr = sns.countplot(x = 'composto', data=compostos_top, order=compostos_top['composto'].value_counts().index) ###Ordenar 
gr.set_title('Top 5')
plt.show()
```
